# **6. Data Pre-Processing**
---

- This section to modify unstructured data to structured data :
  - Handling missing data, typos and inconsistent data.

```
#Filling the missing cities based on Venue of the match was played.
ipl_data.loc[(ipl_data.city.isnull()) & (ipl_data.venue == 'Dubai International Cricket Stadium'), 'city'] ='Dubai'
ipl_data.loc[(ipl_data.city.isnull()) & (ipl_data.venue == 'Sharjah Cricket Stadium'), 'city'] ='Sharjah'

#Convert date feature data type from object to datetime and then extract the year from date as season of the IPL.
ipl_data['date']=pd.to_datetime(ipl_data['date'])
ipl_data['season'] = ipl_data['date'].dt.year

#Filling the 4 missing records in 'player of match', 'winner', 'result' and eliminator features due to matches drawn with 'Match Drawn'.
ipl_data['player_of_match'].fillna('Match Drawn', inplace=True)
ipl_data['winner'].fillna('Match Drawn', inplace=True)
ipl_data['result'].fillna('Match Drawn', inplace=True)
ipl_data['eliminator'].fillna('Match Drawn', inplace=True)

#Filling the missing records in result_margin feature due to match tied with "0".
ipl_data['result_margin'].fillna(0, inplace=True)

#Filling the missing records in method feature as "NA D/L".
ipl_data['method'].fillna('NA D/L', inplace=True)
```

**Q.Which seasons were played the highest number of matches ever?**

```
# setting a figure of size of 20 x 7 inches
figure=plt.figure(figsize=[20,7])

ipl_data['season'].value_counts()
sns.countplot(x='season', data=ipl_data)

# Adding cosmetics to graph
plt.xlabel(xlabel='IPL Season', size=14)
plt.ylabel(ylabel='Number of Matches', size=14)
plt.title(label='IPL Season Wise Number of Matches', size=16)

# Display the plot
plt.show()
```

**Observation: IPL Season 2013 played the 76 maximum number of matches.**

**Q. Which is the most successful team in ipl dataset at hand?**

```
# setting a figure of size of 15 x 7 inches
figure=plt.figure(figsize=[15,7])

most_winner=ipl_data['winner'].value_counts()
sns.barplot(y=most_winner.index, x=most_winner)

# Adding cosmetics to graph
plt.xlabel(xlabel='Match Wining Count', size=14)
plt.ylabel(ylabel='IPL Team Name', size=14)
plt.title(label='Most Successful Team ', size=16)

# Display the plot
plt.show()
```

**Observation: Mumbai Indians and Chennai Super Kings are the most successful teams.**

**Q.What is the probabilities of winning the match if a toss was won?**  

```
toss_factor = ipl_data.toss_winner == ipl_data.winner
toss_factor.value_counts()
```

```
# setting a figure of size of 15 x 7 inches
figure=plt.figure(figsize=[15,7])

toss_data = { "Matches_won_by_toss_winner" : 418,
              "Matches_won_by_toss_looser" : 398}
toss_data = pd.Series(toss_data)
plt.pie(x=toss_data,autopct="%.2f%%",labels=toss_data.index);

# Adding cosmetics to graph
plt.title(label='Win Probability ', size=16)

# Display the plot
plt.show()
```

**Observation: There is a slight differnce for won the toss and won the match.**

**Q. Who is the maximum number of times player of the match?**

```
# setting a figure of size of 15 x 7 inches
figure=plt.figure(figsize=[15,7])

Man_Of_Match=ipl_data['player_of_match'].value_counts()[:10]
sns.barplot(y=Man_Of_Match.index, x=Man_Of_Match)

# Adding cosmetics to graph
plt.xlabel(xlabel='Winner Count', size=14)
plt.ylabel(ylabel='Player Name', size=14)
plt.title(label='Player of the Match Winner', size=16)

# Display the plot
plt.show()
```

**Obersvation: ABD is the most of times player of the match winner. He won 23 times the player of the match awards.**

**Q.In which city were the number of matches played.**

```
# setting a figure of size of 15 x 7 inches
figure=plt.figure(figsize=[15,7])

city=ipl_data['city'].value_counts()
sns.barplot(y=city.index, x=city)

# Adding cosmetics to graph
plt.xlabel(xlabel='Match Count', size=14)
plt.ylabel(ylabel='City Name', size=14)
plt.title(label='Number of Matches played in City', size=16)

# Display the plot
plt.show()
```

**Observation: Mumbai, kolkata and Delhi are the highest match played cities.**

**Q.How many matches were played in neutral venue?**

```
ipl_data['neutral_venue'].value_counts()
```

```
# setting a figure of size of 15 x 7 inches
figure=plt.figure(figsize=[15,7])

venue_data = { "Matches_not_played_in_neutral_venue" : 739,
              "Matches_played_in_neutral_venue" : 77}
venue_data = pd.Series(venue_data)
plt.pie(x=venue_data,autopct="%.2f%%",labels=venue_data.index);

# Adding cosmetics to graph
plt.title(label='Neutral Venue', size=16)

# Display the plot
plt.show()
```

**Q. Who is top 5 umpire in match were palyed?**
